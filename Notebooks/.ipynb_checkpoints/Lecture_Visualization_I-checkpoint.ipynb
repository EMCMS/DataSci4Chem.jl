# Data Visualization

In this lecture/tutorial we will work together on different plotting exercises to be able to represent our data in the best possible manner.
Let's start with first generationg some data into our julia environment. For this we will be using the package *DataSci4Chem.jl*.

## Data Generation

* Generate a vector of values between 1 and 3 (*X*) with a size of 10x1.
* Generate a vector of *Y* by first calculating the sin(X) and then add some noise to those values.

```julia
using DataSci4Chem

X = collect(range(1,3,length=10))  # Generate some x values
Y = sin.(X) .+ rand(10,1)          # Generate some y values
println(X)
println(Y)
```

## E1: 
Let's generate a line plot of the variable *X* and answer the below questions.
    
1. what is plotted in the x axis?
2. what is plotted in the y axis?
3. what is actually being plotted?

```julia
plot(X) 
```

> **Answers to E1**
> 
> 1. The x axis is the index of individual values in the *X* vector.
>
> 2. The actual values of *X* are plotted in in the y domain.
>
> 3. plot(1:length(X),X)

## E2: 
Let's now plot the *X* vs *Y*. 

```julia
plot(X,Y)
```

## E3: 

Let's try to set the x and y label in this plot. How can we do that? 

> **Answers to E3**
> 
> For that there are two different ways of of seeting this up. The first way is by specifying the labels as an attribute in the plot(-) the second way is to use the function xlabel!(-).

```julia
plot(X,Y,xlabel= "x values",ylabel = "y values")  
```

```julia
# or 
plot(X,Y) 
xlabel!("x values") 
ylabel!("y values") 
```
